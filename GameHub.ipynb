```python
#### finallllyyyy

import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk
import random
import pygame
import os

# ---------------- Initialization ----------------
pygame.mixer.init()
# safe-load background music if present
if os.path.exists("background.mp3"):
    try:
        pygame.mixer.music.load("background.mp3")
        pygame.mixer.music.play(-1)
    except Exception as e:
        print("Could not play background.mp3:", e)

# ---------------- User DB & Helpers ----------------
user_db = {}

def check_password_strength(password):
    if len(password) < 6:
        return "Weak"
    elif any(c.isalpha() for c in password) and any(c.isdigit() for c in password):
        return "Medium"
    elif len(password) >= 8 and any(c.isalpha() for c in password) and any(c.isdigit() for c in password) and any(c in "!@#$%^&*()" for c in password):
        return "Strong"
    else:
        return "Weak"

# ---------------- Signup ----------------
def signup_action():
    name = name_entry.get().strip()
    email = email_entry.get().strip()
    password = pass_entry.get()
    confirm_pass = confirm_pass_entry.get()
    sec_q = sec_q_entry.get().strip()
    phone = phone_entry.get().strip()
    dob = dob_entry.get().strip()
    gender = gender_var.get()
    occupation = occ_entry.get().strip()

    if not email or not password or not confirm_pass or not phone:
        messagebox.showerror("Error", "Please fill in all required fields")
        return

    if password != confirm_pass:
        messagebox.showerror("Error", "Passwords do not match")
        return

    strength = check_password_strength(password)
    if strength == "Weak":
        messagebox.showerror("Error", "Password is too weak")
        return

    user_db[email] = {
        "Name": name or "User",
        "Password": password,
        "SecurityQuestion": sec_q,
        "Phone": phone,
        "DOB": dob,
        "Gender": gender,
        "Occupation": occupation
    }

    messagebox.showinfo("Success", f"Account created successfully for {user_db[email]['Name']}")
    print("Database:", user_db)

def signup_frame():
    global name_entry, email_entry, pass_entry, confirm_pass_entry, sec_q_entry, phone_entry, dob_entry, gender_var, occ_entry

    root.withdraw()
    signup_window = tk.Toplevel(root)
    signup_window.title("Signup Page")
    signup_window.geometry("420x440")

    def on_close():
        signup_window.destroy()
        root.deiconify()
    signup_window.protocol("WM_DELETE_WINDOW", on_close)

    tk.Label(signup_window, text="Name").grid(row=0, column=0, padx=10, pady=5, sticky="e")
    name_entry = tk.Entry(signup_window, width=30)
    name_entry.grid(row=0, column=1, padx=10, pady=5)

    tk.Label(signup_window, text="Email ID:").grid(row=1, column=0, padx=10, pady=5, sticky="e")
    email_entry = tk.Entry(signup_window, width=30)
    email_entry.grid(row=1, column=1, padx=10, pady=5)

    tk.Label(signup_window, text="Password:").grid(row=2, column=0, padx=10, pady=5, sticky="e")
    pass_entry = tk.Entry(signup_window, width=30, show="*")
    pass_entry.grid(row=2, column=1, padx=10, pady=5)

    tk.Label(signup_window, text="Confirm Password:").grid(row=3, column=0, padx=10, pady=5, sticky="e")
    confirm_pass_entry = tk.Entry(signup_window, width=30, show="*")
    confirm_pass_entry.grid(row=3, column=1, padx=10, pady=5)

    tk.Label(signup_window, text="Security Question").grid(row=4, column=0, padx=10, pady=5, sticky="e")
    sec_q_entry = tk.Entry(signup_window, width=30)
    sec_q_entry.grid(row=4, column=1, padx=10, pady=5)

    tk.Label(signup_window, text="Phone Number:").grid(row=5, column=0, padx=10, pady=5, sticky="e")
    phone_entry = tk.Entry(signup_window, width=30)
    phone_entry.grid(row=5, column=1, padx=10, pady=5)

    tk.Label(signup_window, text="Date of Birth:").grid(row=6, column=0, padx=10, pady=5, sticky="e")
    dob_entry = tk.Entry(signup_window, width=30)
    dob_entry.grid(row=6, column=1, padx=10, pady=5)

    tk.Label(signup_window, text="Gender:").grid(row=7, column=0, padx=10, pady=5, sticky="e")
    gender_var = tk.StringVar(value="Other")
    tk.Radiobutton(signup_window, text="Male", variable=gender_var, value="Male").grid(row=7, column=1, sticky="w")
    tk.Radiobutton(signup_window, text="Female", variable=gender_var, value="Female").grid(row=7, column=1)
    tk.Radiobutton(signup_window, text="Other", variable=gender_var, value="Other").grid(row=7, column=1, sticky="e")

    tk.Label(signup_window, text="Occupation:").grid(row=8, column=0, padx=10, pady=5, sticky="e")
    occ_entry = tk.Entry(signup_window, width=30)
    occ_entry.grid(row=8, column=1, padx=10, pady=5)

    tk.Button(signup_window, text="Sign Up", width=15, command=signup_action).grid(row=9, column=0, columnspan=2, pady=20)

# ---------------- Login ----------------
def login_action():
    email = login_email_entry.get().strip()
    password = login_pass_entry.get()

    if email not in user_db:
        messagebox.showerror("Error", "No account found with this email.")
        return

    if user_db[email]["Password"] != password:
        messagebox.showerror("Error", "Incorrect password.")
        return

    messagebox.showinfo("Success", f"Welcome back {user_db[email]['Name']}!")
    game_page(login_window)

def login_frame():
    global login_email_entry, login_window, login_pass_entry

    root.withdraw()
    login_window = tk.Toplevel(root)
    login_window.title("Login Page")
    login_window.geometry("320x220")

    def on_close():
        login_window.destroy()
        root.deiconify()
    login_window.protocol("WM_DELETE_WINDOW", on_close)

    tk.Label(login_window, text="Email ID:").grid(row=0, column=0, padx=10, pady=5, sticky="e")
    login_email_entry = tk.Entry(login_window, width=28)
    login_email_entry.grid(row=0, column=1, padx=10, pady=5)

    tk.Label(login_window, text="Password:").grid(row=1, column=0, padx=10, pady=5, sticky="e")
    login_pass_entry = tk.Entry(login_window, width=28, show="*")
    login_pass_entry.grid(row=1, column=1, padx=10, pady=5)

    tk.Button(login_window, text="Login", width=15, command=login_action).grid(row=2, column=0, columnspan=2, pady=20)

# ---------------- Game Selection ----------------
def game_page(login_window):
    login_window.withdraw()
    game_win = tk.Toplevel(root)
    game_win.title("Game Page")
    game_win.geometry("600x700")

    # --- Snake & Ladder Image ---
    try:
        img1 = Image.open("C:/Users/Asus/Downloads/snake and ladder.png")
        img1 = img1.resize((250, 200))
        photo1 = ImageTk.PhotoImage(img1)
    except Exception:
        photo1 = None

    if photo1:
        label1 = tk.Label(game_win, image=photo1)
        label1.image = photo1
        label1.grid(row=0, column=0, padx=20, pady=20)
    else:
        tk.Label(game_win, text="Snake & Ladder").grid(row=0, column=0, padx=20, pady=20)

    btn1 = tk.Button(game_win, text="Play Snake & Ladder", width=20, height=2, command=start_snake_ladder)
    btn1.grid(row=1, column=0, pady=10)

    # --- Tic Tac Toe Image ---
    try:
        img2 = Image.open("C:/Users/Asus/Downloads/tic tac toe.png")
        img2 = img2.resize((250, 200))
        photo2 = ImageTk.PhotoImage(img2)
    except Exception:
        photo2 = None

    if photo2:
        label2 = tk.Label(game_win, image=photo2)
        label2.image = photo2
        label2.grid(row=0, column=1, padx=20, pady=20)
    else:
        tk.Label(game_win, text="Tic Tac Toe").grid(row=0, column=1, padx=20, pady=20)

    btn2 = tk.Button(game_win, text="Play Tic Tac Toe", width=20, height=2, command=start_tic_tac_toe)
    btn2.grid(row=1, column=1, pady=10)

# ---------------- Tic Tac Toe (in-window) ----------------
def start_tic_tac_toe():
    # Create a new window for tic tac toe
    ttt_win = tk.Toplevel(root)
    ttt_win.title("Tic Tac Toe")
    cell_size = 100
    ttt_win.geometry(f"{cell_size*3}x{cell_size*3}")

    current_player = {"val": "X"}  # use dict to allow inner functions to mutate
    grid = [[ "" for _ in range(3)] for _ in range(3)]
    canvas_grid = [[ None for _ in range(3)] for _ in range(3)]

    def check_winner():
        for row in range(3):
            if grid[row][0] == grid[row][1] == grid[row][2] != "":
                return True
        for col in range(3):
            if grid[0][col] == grid[1][col] == grid[2][col] != "":
                return True
        if grid[0][0] == grid[1][1] == grid[2][2] != "":
            return True
        if grid[0][2] == grid[1][1] == grid[2][0] != "":
            return True
        return False

    def check_tie():
        for row in range(3):
            for col in range(3):
                if grid[row][col] == "":
                    return False
        return True

    def draw_glow(canvas):
        canvas.create_oval(20, 20, 80, 80, fill="#66ccff", outline="")
        canvas.create_oval(25, 25, 75, 75, fill="#33aaff", outline="")
        canvas.create_oval(30, 30, 70, 70, fill="black", outline="")

    def draw_X(canvas):
        canvas.create_line(20, 20, 80, 80, fill="#ff99ff", width=10)
        canvas.create_line(20, 80, 80, 20, fill="#ff99ff", width=10)
        canvas.create_line(25, 25, 75, 75, fill="#ff33cc", width=6)
        canvas.create_line(25, 75, 75, 25, fill="#ff33cc", width=6)
        canvas.create_line(30, 30, 70, 70, fill="#ff00ff", width=2)
        canvas.create_line(30, 70, 70, 30, fill="#ff00ff", width=2)

    def on_click(row, col):
        if grid[row][col] == "":
            grid[row][col] = current_player["val"]
            canvas = canvas_grid[row][col]
            if current_player["val"] == "X":
                draw_X(canvas)
            else:
                draw_glow(canvas)

            if check_winner():
                pygame.mixer.music.stop()
                messagebox.showinfo("Game Over", f"Player {current_player['val']} wins!")
                ttt_win.destroy()
                # Optionally restart background music
                try:
                    pygame.mixer.music.play(-1)
                except Exception:
                    pass
            elif check_tie():
                pygame.mixer.music.stop()
                messagebox.showinfo("Game Over", "It's a tie!")
                ttt_win.destroy()
                try:
                    pygame.mixer.music.play(-1)
                except Exception:
                    pass
            else:
                current_player["val"] = "O" if current_player["val"] == "X" else "X"

    # Create grid canvases
    for r in range(3):
        for c in range(3):
            canvas = tk.Canvas(ttt_win, width=cell_size, height=cell_size, bg="black", highlightthickness=1, highlightbackground="white")
            canvas.grid(row=r, column=c)
            canvas.bind("<Button-1>", lambda e, rr=r, cc=c: on_click(rr, cc))
            canvas_grid[r][c] = canvas

# ---------------- Snake & Ladder ----------------
def start_snake_ladder():
    snl_win = tk.Toplevel(root)
    snl_win.title("Snake & Ladder")
    cell_size = 60
    board_size = 10
    snl_win.geometry(f"{cell_size*board_size+20}x{cell_size*board_size+140}")

    canvas = tk.Canvas(snl_win, width=cell_size*board_size, height=cell_size*board_size)
    canvas.pack(pady=10)

    # Load board image (if available), else keep blank board
    try:
        img = Image.open("C:/Users/Asus/Downloads/Photo.jpg")
        img = img.resize((cell_size*board_size, cell_size*board_size))
        photo = ImageTk.PhotoImage(img)
        canvas.photo = photo
        canvas.create_image(0, 0, anchor="nw", image=photo)
    except Exception:
        # draw cells as fallback
        for r in range(board_size):
            for c in range(board_size):
                x0 = c*cell_size
                y0 = r*cell_size
                canvas.create_rectangle(x0, y0, x0+cell_size, y0+cell_size, outline="gray")

    snakes = {17: 7, 54: 34, 62: 19, 64: 60, 87: 36, 93: 73, 95: 75, 98: 79}
    ladders = {1: 38, 4: 14, 9: 31, 21: 42, 28: 84, 51: 67, 72: 91, 80: 99}

    players = {"Player 1": {"pos": 1, "color": "red"}, "Player 2": {"pos": 1, "color": "blue"}}
    tokens = {}
    offsets = {"Player 1": -15, "Player 2": 15}

    # initial tokens at start
    for player, info in players.items():
        x, y = 20, (board_size-1)*cell_size + 20
        tokens[player] = canvas.create_oval(x-10, y-10, x+10, y+10, fill=info["color"])

    turn_order = list(players.keys())
    turn_index = 0

    def get_coords(pos):
        """Get canvas coordinates for a given board position"""
        pos = max(1, min(100, pos))
        row = (pos-1)//10
        col = (pos-1)%10
        if row % 2 == 1:  # zigzag
            col = 9-col
        x = col*cell_size + cell_size//2
        y = (board_size-1-row)*cell_size + cell_size//2
        return x, y

    def move_player(player, steps):
        pos = players[player]["pos"] + steps
        if pos > 100:
            pos = players[player]["pos"]
        if pos in snakes:
            messagebox.showinfo("Snake üêç", f"Oh no! {player} bitten by a snake at {pos}")
            pos = snakes[pos]
        elif pos in ladders:
            messagebox.showinfo("Ladder ü™ú", f"Yay! {player} climbed a ladder from {pos}")
            pos = ladders[pos]
        players[player]["pos"] = pos

        x, y = get_coords(pos)
        canvas.coords(tokens[player], x-10+offsets[player], y-10, x+10+offsets[player], y+10)

        if pos == 100:
            messagebox.showinfo("Winner üéâ", f"{player} Wins the Game!")
            pygame.mixer.music.stop()
            snl_win.destroy()

    dice_label = tk.Label(snl_win, text="üé≤ Roll the dice!", font=("Arial", 14))
    dice_label.pack(pady=8)

    def roll_dice():
        nonlocal turn_index
        player = turn_order[turn_index % len(players)]
        steps = random.randint(1, 6)
        dice_label.config(text=f"{player} rolled a {steps}")
        move_player(player, steps)
        turn_index += 1

    roll_btn = tk.Button(snl_win, text="üé≤ Roll Dice", font=("Arial", 14), bg="orange", command=roll_dice)
    roll_btn.pack(pady=10)

# ---------------- Main Page ----------------
root = tk.Tk()
root.title("Main Page")
root.geometry("500x600")

top_frame = tk.Frame(root, height=40)
top_frame.pack(side="top", fill="x")

signup_btn = tk.Button(top_frame, text="Sign Up", command=signup_frame)
signup_btn.pack(side="right", padx=10, pady=5)

login_btn = tk.Button(top_frame, text="Login", command=login_frame)
login_btn.pack(side="right", padx=10, pady=5)

# frontpage image
try:
    img = Image.open("C:/Users/Asus/Downloads/frontpage.png")
    img = img.resize((450, 300))
    photo = ImageTk.PhotoImage(img)
    label = tk.Label(root, image=photo)
    label.image = photo
    label.place(relx=0.5, rely=0.5, anchor="center")
except Exception:
    tk.Label(root, text="Welcome!", font=("Arial", 24)).place(relx=0.5, rely=0.5, anchor="center")

root.mainloop()
```
